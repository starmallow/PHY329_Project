```python
import matplotlib.pyplot as plt

timesteps = 100
cars = 20
cells = 100

bn_start = 40
bn_end   = 60
v_max_bn = 1
inflow   = 0.5

test_bn = TrafficBottleneck(cars=cars, cells=cells, bn_start=bn_start, bn_end=bn_end, v_max_bn=v_max_bn, inflow=inflow)
test_bn.simulate(timesteps)
data_bn = test_bn.history

fig, ax = plt.subplots(figsize=(12, timesteps//10), dpi=300)
im = ax.imshow(data_bn, cmap='Greys')

# ax.set_xticks([])
# ax.set_yticks([])
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)

ax.set_ylabel("Timestep", fontsize=12)
ax.set_xlabel("Cell Index", fontsize=12)
ax.set_title("Time Evolution of Traffic With Bottleneck", fontsize=14)

cbar = plt.colorbar(im, ax=ax, pad=0.02)
cbar.set_label(
    "Car Velocity (cells/timestep)\n(-1 indicates no cars present)",
    fontsize=10,
    labelpad=1
)

plt.show()
```
