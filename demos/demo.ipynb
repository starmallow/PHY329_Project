# Visualizing the evolution of the traffic lane cells over time

The `SpaceTimePlot` function streamlines the creation of time evolution plots of the circular and bottleneck Nagelâ€“Schreckenberg models. This function defaults to modeling the circular traffic system, and will switch to the bottleneck model by using the argument: `bottleneck = True`.

## Circular Model System Density:

Varying the `system_density` argument demonstrates the transition from smooth laminar flow to considerable traffic congestion.

### Laminar flow at low density:

```python
SpaceTimePlot(cells=100, system_density=0.05, timesteps=100)
```

### Transitioning traffic flow with increasing density:

```python
SpaceTimePlot(cells=100, system_density=0.1, timesteps=100)
SpaceTimePlot(cells=100, system_density=0.12, timesteps=100)
SpaceTimePlot(cells=100, system_density=0.14, timesteps=100)
```

### Traffic congestion at higher densities:

```python
SpaceTimePlot(cells=100, system_density=0.2, timesteps=100)
SpaceTimePlot(cells=100, system_density=0.3, timesteps=100)
SpaceTimePlot(cells=100, system_density=0.5, timesteps=100)
```
